```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn
import scipy as sp 
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
import os
# import sys; sys.path.append('/home/mfc/Dropbox/spring2016/dataMining/projects/project2/')
root_dir = '/home/mfc/Dropbox/spring2016/dataMining/projects/project2/'
os.chdir(root_dir)
import numpy as np

from kcenter import *


%matplotlib inline
```

```python
data = pd.read_csv('data/wdbc.csv')
data.head(5)
```

```python
X = data.iloc[:,2:].values


# normalizing the data
X = scale(X)
```

```python
pca = PCA(n_components=2)

X_2 = pca.fit_transform(X)
```

```python

plt.figure(figsize=(10,10))
plt.scatter(X_2[:,0], X_2[:,1], c=data.iloc[:,1], s=50, alpha=0.6)
plt.xlabel('PCA_1')
plt.ylabel('PCA_2')
```

```python
c,r,label = kcenter(X_2, 2)


ax = plt.figure(figsize=(15,15))
plt.scatter(X_2[:,0], X_2[:,1], c=label, 
            cmap=plt.cm.gnuplot, s=50, alpha=0.7)


plt.scatter(c[0][0],c[0][1], s=400)
plt.scatter(c[1][0],c[1][1], s=400)

# plt.xlim((-12,12))
# plt.ylim((-12,12))

circle1 = plt.Circle((c[0][0],c[0][1]),
                     r[0], color='b',
                     fill=False, clip_on=True, alpha=1)
ax.gca().add_artist(circle1)

circle2 = plt.Circle((c[1][0],c[1][1]),
                     r[1], color='g',
                    fill=False, clip_on=True, alpha=1)
ax.gca().add_artist(circle2)


plt.xlabel('PCA_1')
plt.ylabel('PCA_2')

```
