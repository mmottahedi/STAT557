```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn
import scipy as sp 
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
import os
import numpy as np
root_dir = '/home/mfc/Dropbox/spring2016/dataMining/projects/project2/'
os.chdir(root_dir)

from kcenter import *


%matplotlib inline
```

```python


data = pd.read_csv('data/breast-cancer-wisconsin.csv',
                   names=['id','thickness','size.unif','shape.unif','adhesion',
                          'size.cell','nuclei','chromatin','nucleoli','mitoses','class'],
        na_values='?')

data = data.dropna().copy()
```

```python
X = data.iloc[:,1:].values



# normalizing the data
X = scale(X)
```

```python
pca = PCA(n_components=2)

X_2 = pca.fit_transform(X)
```

```python

plt.figure(figsize=(10,10))
plt.scatter(X_2[:,0], X_2[:,1], c=data.iloc[:,-1], s=50, alpha=0.6, cmap=plt.cm.gnuplot)
plt.xlim((-8,4))
plt.ylim((-4,6))
plt.xlabel('PCA_1')
plt.ylabel('PCA_2')
```

```python
c,r,label = kcenter(X_2, 2)


ax = plt.figure(figsize=(7,4))
plt.scatter(X_2[:,0], X_2[:,1], c=label, 
            cmap=plt.cm.gnuplot, s=50, alpha=0.7)


plt.scatter(c[0][0],c[0][1], s=400)
plt.scatter(c[1][0],c[1][1], s=400)

plt.xlim((-8,3))
plt.ylim((-3,5))

circle1 = plt.Circle((c[0][0],c[0][1]),
                     r[0], color='black',
                     fill=0, clip_on=True, alpha=0.1)
ax.gca().add_artist(circle1)

circle2 = plt.Circle((c[1][0],c[1][1]),
                     r[1], color='yellow',
                     fill=0, clip_on=True, alpha=0.1)
ax.gca().add_artist(circle2)


plt.xlabel('PCA_1')
plt.ylabel('PCA_2')

plt.savefig('./doc/m1.png')

```
