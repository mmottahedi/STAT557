```python
from LDA import DA
import numpy as np
from ggplot import *
from pylab import *
%matplotlib inline



np.random.seed(1234)
cov1 = [[2, 5], [0, 1]]
cov2 = [[1, 1], [0, 1]]
mean1 = [0, 0]
mean2 = [-3, -2]
Y = np.random.randint(0, 2, 1000)
X1 = np.random.multivariate_normal(mean1, cov1, 475)
X2 = np.random.multivariate_normal(mean2, cov2, 525)
X = np.zeros((1000,2))
X[Y == 0] = X1
X[Y == 1] = X2

fit = DA(X, Y, mode="QDA")
fit.sampleEstimate()
yhat = fit.predict(X)
fit.summary()
```

```python
cov1 = [[2, 5], [0, 1]]
cov2 = [[1, 1], [0, 1]]
mean1 = [0, 0]
mean2 = [-4, -2]
Y = np.random.rand(1000)
X1 = np.random.multivariate_normal(mean1, cov1, 500)
X2 = np.random.multivariate_normal(mean2, cov2, 500)
X = np.zeros((1000,3))
sum(Y==1)

X[:500,:2] = X1
X[500:, :2] = X2
X[:, 2] = Y


scatter(X1[:,0],X1[:,1],color='r', alpha = 0.3)
scatter(X2[:,0],X2[:,1],color='b', alpha = 0.3)

```

```python
from scipy.linalg import svd, eig
from sklearn.decomposition import PCA
X -= X.mean(axis=0)
pca = PCA(n_components=2)
pca.fit(X)
print(pca.explained_variance_ratio_)
u,d,v = svd(X, full_matrices=False)
eig(np.dot(X.T,X))
```
