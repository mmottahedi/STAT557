```python
from LDA import DA
import numpy as np
from ggplot import *
from pylab import *
%matplotlib inline


np.random.seed(1234)
cov1 = [[2, 0], [0, 2]]
cov2 = [[2, 0], [0, 2]]
mean1 = [0, 0]
mean2 = [-2, -2]
Y = np.random.randint(0, 2, 1000)
X1 = np.random.multivariate_normal(mean1, cov1, 475)
X2 = np.random.multivariate_normal(mean2, cov2, 525)
X = np.zeros((1000,2))
X[Y == 0] = X1
X[Y == 1] = X2

fit = DA(X, Y, mode="QDA")
fit.sampleEstimate()
yhat = fit.predict(X)
fit.summary()
```
