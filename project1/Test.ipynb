```python
from LDA import DA
import numpy as np

np.random.seed(1234)
cov1 = [[1.7, -0.14], [-.14, 1.6]]
cov2 = [[2, 0], [0, 2]]
mean1 = [2, 2]
mean2 = [-2, -2]
Y = np.random.randint(0, 2, 1000)
X1 = np.random.multivariate_normal(mean1, cov1, 475)
X2 = np.random.multivariate_normal(mean2, cov2, 525)
X = np.zeros((1000,2))
X[Y == 0] = X1
X[Y == 1] = X2

fit = DA(X, Y, mode="QDA")
fit.sampleEstimate()
yhat = fit.predict(X)
print(fit.xDim, fit.yDim)
print(fit.classMean)
print(fit.priorProbability)
print("common covariance matrix: \n",fit.commonCovMatrix)
print(type(fit.commonCovMatrix))
fit.summary()
```
